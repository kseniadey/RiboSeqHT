# Задание 2 . Анализ wig-файла

0. Чтение wig-файлов

```python
def wig_parser(file, genome_size):
    counts = np.zeros(genome_size)
    with open(file, 'r') as file:
        file.readline()
        for line in file:
            data = line.split()
            position, count = int(data[0]), int(data[1])
            position = position - 2
            counts[position] = count
        return counts
```

```python
forward = wig_parser('02. RiboSeq_EColi_track_f.wig', 4641668)
print(forward)
```

```python
reverse = wig_parser('02. RiboSeq_EColi_track_r.wig', 4641668)
print(reverse)
```

```python
from BCBio import GFF

from_gb = "sequence.gff3"

limit_info = dict(gff_type = ["CDS"])

genbank = open(from_gb)
seq_records = list(GFF.parse(genbank, limit_info = limit_info))
genbank.close()
```

1. Присутствует ли в покрытии генома неравномерность? Рассмотрите неравномерность покрытия на примере любого выбранного вами гена. Что вы можете сказать про эту неравномерность? Присутствует ли там периодичность в три нуклеотида? А если взять среднее по большому количеству генов (взяв за точку отсчёта старт-кодон)? Выводы проиллюстрируйте графиками.

На больших участках (1000<) не удалось проследить периодичность, для наглядности взяла маленький участок

```python
SeqFeatures = []
for feature in seq_records[0].features:
    if feature.type == "CDS": 
        SeqFeatures += [feature]
        
mapped = {}
mapped_cov = np.zeros(200, dtype = np.float64) 
for feature in SeqFeatures:
    start, end = feature.location.nofuzzy_start, feature.location.nofuzzy_end
    length = end - start
    if length % 3 == 0:
        try:
            strand_count = forward if feature.strand == 1 else reverse
        except:
                pass   
    mapped[feature.id] = sum(strand_count[start:end])
    if length >= mapped_cov.size:
        mapped_cov = mapped_cov + strand_count[start:start + mapped_cov.size]
```

```python
position = np.arange(mapped_cov.size)
fig = plt.figure(figsize = (20,10))
ax = fig.add_subplot(111)
ax.bar(position, mapped_cov, color = ["fuchsia" if nt % 3 == 1 else "black" for nt in position])
ax.set_title("Periods of counts")
ax.set_xlabel("Position")
ax.grid()
```

На начальных позициях периодичность в 3 нуклеотида не наблюдается,однако, чем дальше позиция, видно, что каждый 3 третий нуклеотид показывает более высокое значение. Также прослеживается что каждый ~ 40 пик (розовый) выше чем все остальные. Интересно, что саааамый высокий пик не на первом кодоне, а на втором, вероятно, рибосома застряла именно здесь.

### 2. Рассчитайте значения RPKM и TPM для каждого гена. В чём разница между RPKM и TPM? Для чего используется каждая метрика?

Данные метрики используются для нормирования глубины прочтения и длинны гена. 

Принципиальное отличие в том, что при подсчете ТРМ сначала производится нормирование для длинны и только после нормирование для глубины прочтения. Сумма всех ТРМ в каждом из образцов одинакова,что облегчает сравнение доли прочтений, картированных на геном в образцах. В то время как RPKM -- сумма нормированных ридов отличны от образца к образцу, что приводит к тому что данная метрка не сопоставима в различных образцах. Таким образом, с помощью TPM можно сравнить между собой несколько образцов.

```python
TPM = {}
RPKM = {}

for feature in SeqFeatures:
    start, end = feature.location.nofuzzy_start, feature.location.nofuzzy_end
    length = end - start
    if length % 3 == 0:
        try:
            RPKM[feature.id] = mapped[feature.id] * 10**9 / (sum(mapped.values()) * length)
            TPM[feature.id] = 10**6 * RPKM[feature.id]/sum(RPKM.values())
        except:
            pass 
```
